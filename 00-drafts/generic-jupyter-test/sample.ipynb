# Generate synthetic data
Let's create a simple relationship: y = 3x + 5 with some added noise

```python
import numpy as np

np.random.seed(42)  # for reproducible results
X = 2 * np.random.rand(100, 1)  # 100 samples, single feature
y = 3 * X + 5 + np.random.randn(100, 1)  # true slope=3, intercept=5
```

# Create and train the model


```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()      # instantiate the model
model.fit(X, y)                 # train (fit) the model on data
```

# Model parameters


```python
print("Estimated slope (coefficient):", model.coef_[0][0])
print("Estimated intercept:", model.intercept_[0])
```

# Make predictions

```python
X_new = np.array([[0], [2]])    # for example, predict at X=0 and X=2
y_pred = model.predict(X_new)
print("Predictions for X=0 and X=2:", y_pred.ravel())
```

# Plot the results


```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Model predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression with scikit-learn')
plt.legend()
plt.show()
```
